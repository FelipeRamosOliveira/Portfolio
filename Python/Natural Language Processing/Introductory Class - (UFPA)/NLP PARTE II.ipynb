# ESTRUTURANDO DADOS NÃO ESTRUTURADOS



**INFORMAÇÃO x DADO**

Antes de entrarmos de fato na definição da estrutura dos dados, você deve saber o que são dados. Em resumo, um dado é a representação utilizada para armazenar uma informação (textos,fotos,videos,tabelas, por exemplo).

![](https://www.oriresults.com/wp-content/uploads/Blog-Whats-Hiding-in-Your-Unstructured-Data-1000x592px.png)

**DADOS ESTRUTURADOS**

Como o próprio nome diz, os dados estruturados são aqueles que possuem estruturas bem definidas, rígidas, pensadas antes da própria existência do dado que será carregado naquela estrutura.

Não permitem que tipos de dados diferentes das estruturas pré-estabelecidas sejam carregados. Por exemplo, se a coluna de uma tabela foi criada para ser numérica, ela não aceitará dados textuais.

Em sua maioria, dentro de seus sistemas, existem bancos de dados criados diretamente pelos usuários de ambientes e tecnologias analíticas (arquivos Excel, por exemplo). 

**DADOS NÃO ESTRUTURADOS**

Não possuem estruturas bem definidas, alinhadas, padronizadas, podendo ser compostos por diversos elementos diferentes dentro um todo (milhares de pixels únicos dentro de uma mesma foto de uma paisagem, por exemplo).

Dados não estruturados são a forma mais frequente de representação de informação, exatamente porque são criados a partir do uso cotidiano das tecnologias na sociedade (mensagens de texto, selfies, áudios via WhatsApp, entre outros).

Exemplos: fotos, áudios, vídeos, textos livres, emojis e stickers 

# 3.ANÁLISE DE FREQUÊNCIA E A BIBLIOTECA PANDAS 

O Pandas é um biblioteca para manipulação e análise de dados, escrita em Python. Essa é a biblioteca perfeita para iniciar suas análises exploratórias de dados, pois ela nos permite ler, manipular, agregar e plotar os dados em poucos passos.

## 3.1. Tabelar informação

```python
# Importar pandas
import pandas as pd

# Importar colletions
from collections import Counter
```

```python
# Ler um arquivo de texto
rap_lord=open('Rap Lord.txt','r',encoding="utf8").read()
```

```python
#Criar uma tabela
tabela=pd.DataFrame(Counter(rap_lord.split()).most_common(10000))
tabela
```

```python
#Renomear colunas
tabela.rename(columns={0:"Palavra",1:"Contagem"},inplace=True)
tabela
```

```python
#Total de palavras
total=sum(Counter(rap_lord.split()).values())
#Criar coluna de frequências 
tabela["Percentual"]=(tabela["Contagem"]/total)*100
#Visualizar
tabela
```

## 3.1. Visualizar dados (*Seaborn*)

Seaborn é uma biblioteca de visualização de dados Python baseada em matplotlib. Ele fornece uma interface de alto nível para desenhar gráficos estatísticos atraentes e informativos.

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(15,5)})
```

```python
ax = sns.barplot(x="Palavra", y="Percentual",palette='Greens_r', data=tabela.head(20))
```

# 4.NORMALIZAÇÃO DE TEXTOS 

A normalização de texto é o processo de transformar o texto em uma única forma canônica . A normalização do texto antes de armazená-lo ou processá-lo permite a separação de prioridades, uma vez que a entrada é garantida como consistente antes que as operações sejam executadas nele.

## 4.1. Estabelecer um padrão para as letras (maiúsculas ou minusculas)

```python
#Aplicar padrão
rap_lord=rap_lord.lower()
rap_lord
```

##4.2. Remover acentos

```python
#Instalar biblioteca de simbolos
! pip install Unidecode
```

```python
#Chamar a biblioteca instalada
import unidecode
```

```python
#Remover acentos
rap_lord=unidecode.unidecode(rap_lord)
rap_lord
```

## 4.3. Remover caracteres especiais

```python
# Importar biblioteca de Regex
import re
```

```python
#Remover caracteres especiais
rap_lord=re.sub('\W+',' ',rap_lord)
rap_lord
```

## 4.4. Teste da seção
Crie um gáfico de frequência das 20 palavras mais comuns no texto, pós normalização.

```python
#Formar tabela
tabela=pd.DataFrame(Counter(rap_lord.split()).most_common(10000))
#Renomear colunas
tabela.rename(columns={0:"Palavra",1:"Contagem"},inplace=True)
#Total de palavras
total=sum(Counter(rap_lord.split()).values())
#Criar coluna de frequências 
tabela["Percentual"]=(tabela["Contagem"]/total)*100
#Visualizar
tabela.head(10)
```

```python
ax = sns.barplot(x="Palavra", y="Percentual",palette='Greens_r', data=tabela.head(20))
```

```python
# Chamar a biblioteca
import nltk
# Instalar modulos em português
nltk.download('stopwords')
```

## 5.2. Remover palavras vazias (*stop words*)

Palavras irrelevantes (stop words) são palavras em inglês que não acrescentam muito significado a uma frase. Eles podem ser ignorados com segurança sem sacrificar o significado da frase. Por exemplo, as palavras como "de", "da","que", etc.

```python
#Chamar biblioteca de stop words do nltk
stopwords = nltk.corpus.stopwords.words('portuguese')
#Visulizar exemplos
stopwords[:10]
```

```python
#Remover
palavras_sem_stopwords = [palavra for palavra in rap_lord.split() if palavra not in stopwords]
```

```python
#Visualizar em forma de tabela
tabela=pd.DataFrame(Counter(palavras_sem_stopwords).most_common(10000))
tabela
```

## 5.3. Teste da seção
Crie um gáfico de frequência das 20 palavras mais comuns no texto, pós remoção da stop words.

```python
#Formar tabela
pd.DataFrame(Counter(palavras_sem_stopwords).most_common(10000))
#Renomear colunas
tabela.rename(columns={0:"Palavra",1:"Contagem"},inplace=True)
#Total de palavras
total=sum(Counter(palavras_sem_stopwords).values())
#Criar coluna de frequências 
tabela["Percentual"]=(tabela["Contagem"]/total)*100
#Visualizar
tabela.head(10)
```

```python
ax = sns.barplot(x="Palavra", y="Percentual",palette='Greens_r', data=tabela.head(20))
```
